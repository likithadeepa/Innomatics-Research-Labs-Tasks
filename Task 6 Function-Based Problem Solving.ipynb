```python
#Smart Parking Lot Management System
def smart_parking(capacity,logs):
    parked=0
    peak=0
    for log in logs:
        if log=="in":
            parked += 1
        elif log=="out" and parked > 0:
            parked-=1
        if parked>peak:
            peak=parked
    print("currently Parked Vehicle:",parked)
    if parked > capacity:
        print("Parking status:Full - Alert!")
    else:
        print("parking status:Available")

smart_parking(50,["in","in","in", "out","in","in","out"])
```

```python
#Online Food Delivery Time Estimator
def estimate_delivery_time(distance,traffic,weather):
    base_time =distance*5  
    if traffic == "high":
        base_time += 15
    elif traffic=="medium":
        base_time += 8
    if weather=="rainy":
        base_time += 10
    elif weather== "stormy":
        base_time+= 20
    print("Estimated delivery time:",base_time, "minutes")
estimate_delivery_time(8,"high","rainy")
```

```python
#Movie Theatre Seat Occupancy Analyzer
def seat_occupancy(total_seats,booked_seats):
    booked_count=len(booked_seats)
    occupancy =(booked_count/total_seats)*100
    print("occupancy:",int(occupancy),"%")
    if occupancy==100:
        print("Show status:housefull")
    elif occupancy >= 75:
        print("Show status:full")
    else:
        print("Show status:Seats Available")
seat_occupancy(200,[1]*150)
```

```python
#Cloud Server Load Classification System
def classify_server_load(cpu_readings):
    total=0
    for reading in cpu_readings:
        total += reading
    avg_cpu = total / len(cpu_readings)
    print("Average CPU load:",int(avg_cpu), "%")
    if avg_cpu < 50:
        print("Server status:Normal")
    elif 50 <= avg_cpu <= 80:
        print("Server status:Warning")
    else:
        print("Server status:Critical")
classify_server_load([45,60,70,85,90])
```

```python
#Smart Classroom Resource Usage Monitor
def classroom_monitor(resources):
    overused=[]
    for resource, hours in resources.items():
        if hours> 8:  
            overused.append(resource)
    print("Overused Resources:", ", ".join(overused))
    if overused:
        print("Energy alert:yes")
    else:
        print("Energy alert:no")
usage ={
    "projector":6,
    "AC":9,
    "ights":4
}
classroom_monitor(usage)
```

```python
#Online Event Registration Capacity Controller
def event_registration(capacity,registrations):
    if registrations>capacity:
        confirmed =capacity
        waitlisted=registrations-capacity
        status="closed"
    else:
        confirmed=registrations
        waitlisted=0
        status="Open"
    print("confirmed registrations:",confirmed)
    print("waitlisted users:",waitlisted)
    print("registration status:",status)
event_registration(100,105)
```
