```python
#Pass/Fail Analyzer
marks=[45,78,90,33,60]
pass_count=0
fail_count=0
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1
print("total students:",len(marks))
print("number of pass students:",pass_count)
print("Number of fail students:",fail_count)
if pass_count>fail_count:
    print("Overall Result: majority passed")
elif fail_count>pass_count:
    print("Overall result:majority failed")
else:
    print("overall result:Equal Pass and Fail")
```

```python
#Simple Data Cleaner
names=["alice","bob","CHARLIE"]
cleaned_names=[]
for name in names:
    cleaned_name=name.strip().lower()
    cleaned_names.append(cleaned_name)
print("original list:",names)
print("cleaned list:",cleaned_names)
```

```python
#Message Length analyzer 
messages=["hi","welcome to the platform","ok"]
for msg in messages:
    length=len(msg)
    print(f"Message:'{msg}'|Length: {length}")
    if length>10:
        print("status:long message")
    else:
        print("status:short message")
    print()
```

```python
# Error Message Detector
logs=["info","error","warning","error"]
error_count=0
for log in logs:
    if log=="error":
        error_count +=1
print("total error entries:",error_count)
```
