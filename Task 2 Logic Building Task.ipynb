```python
#unique word in sentence 
sentence = input("enter sentence: ")
words = sentence.split()
unique_words = set(words)
print("unique words count:", len(unique_words))
print("unique words:", unique_words)
```

```python
# highest salary from employee data
employee={
    "ravi":75000,
    "anita":68000,
    "kiran":72000
}
highest_name=""
highest_salary=0
for name,salary in employee.items():
    if salary>highest_salary:
        highest_salary=salary
        highest_name=name
print("highest Salary:",highest_name,"-",highest_salary)
```

```python
#find aximum and minimum values
n=[45,22,89,10,66]
max_value=n[0]
min_value=n[0]
for num in n:
    if num>max_value:
        max_value=num
    if num<min_value:
        min_value=num
print("list:",n)
print("max:",max_value)
print("min:",min_value)
```

```python
#Calculate attendance percentage
attendance=["b", "a", "p", "a", "p","p"]
total_days=len(attendance)
present_days=attendance.count("p")
percentage=(present_days/total_days) * 100
print("attendance percentage:", percentage)
```

```python
#Count character frequency
text="pythonp"
char_count={}
for char in text:
    if char in char_count:
        char_count[char]+= 1
    else:
        char_count[char] =1
print(char_count)
```
