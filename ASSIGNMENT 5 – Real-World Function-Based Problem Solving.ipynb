```python
#Social Media – Post Engagement Analyzer
def analyze_engagement(likes_list):
    total_likes=0
    for like in likes_list:
        total_likes+=like
    if total_likes >=1000:
        status="viral Post"
    else:
        status ="normal engagement"
    print("total likes:",total_likes)
    print("Post status:",status)
analyze_engagement([200,300,150,400])
```

```python
#Healthcare – Medicine Stock Alert System
def check_medicine_stock(stock):
    if stock<10:
        status="low stock alert"
    else:
        status="Stock Sufficient"
    print("medicine stock:",stock)
    print("status:",status)
check_medicine_stock(6)
```

```python
#Agriculture Rainfall Adequacy Checker
def rainfall_checker(rainfall_list,required_level):
    total=0
    for rain in rainfall_list:
        total+=rain
    average =total /len(rainfall_list)
    if average>=required_level:
        status="Adequate rainfall"
    else:
        status="inadequate rainfall"
    print("average rainfall:", average)
    print("rainfall status:", status)
rainfall_checker([60,70,80,78],70)
```

```python
#social MediaDuplicate Account Detection
def detect_duplicates(usernames):
    unique_users =set(usernames)
    if len(unique_users)!=len(usernames):
        result="yes"
    else:
        result="no"
    print("Duplicate accounts found:",result)
detect_duplicates(["user1","user2","user1","user3"])
```

```python
# Healthcare Appointment Eligibility Checker
def appointment_eligibility(age):
    if age >=18:
        status="eligible"
    else:
        status="not Eligible"
    print("Patient age:", age)
    print("Eligibility status:",status)
appointment_eligibility(21)
```

```python
#agriculture Premium Crop Price Filter
def premium_crop_filter(prices):
    premium=[]
    for price in prices:
        if price>2000:
            premium.append(price)
    print("premium crops:",premium)
premium_crop_filter([1500,2500,1800,3200])
```

```python
#System Monitoring Application Health Checker
def application_health(errors):
    if errors==0:
        status="healthy"
    elif errors <= 5:
        status="Minor issues"
    else:
        status="critical issues"
    print("Error count:",errors)
    print("System status:",status)
application_health(7)
```

```python
#E-Learning – Student Attendance Eligibility System
def attendance_checker(attended_classes, total_classes):
    percentage=(attended_classes /total_classes)*100
    if percentage>=75:
        status="eligible"
    else:
        status="not eligible"
    print("Attendance percentage:",percentage)
    print("Exam Eligibility:", status)
attendance_checker(40,50)
```

```python
#Smart Electricity Bill Analyzer
def electricity_bill(units):
    if units <= 100:
        bill =units*3
    elif units <= 200:
        bill=(100*3)+(units-100)*5
    else:
        bill=(100*3)+(100*5)+(units-200)*7
    if bill<500:
        usage="Low Usage"
    elif 500<=bill<=1500:
        usage ="moderate usage"
    else:
        usage ="High Usage"
    return bill, usage
bill_amount, status =electricity_bill(250)
print("Total Bill:", bill_amount)
print("Usage Level:", status)
```
